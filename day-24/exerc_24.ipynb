```python
array1 = np.array([[1,2,3,4]]) #2d array vain yhdellä arvolla(lista) 1 row, 4 column
array2 = np.array([[1],[2],[3],[4]]) # 2d array jossa 4 row, 1 column

print(array1.shape)
print(array2.shape)
```

```python
#kokeillaan voidaanko tehdä broadcast, eli tehdä aritmeettisia operaatioita taulukoiden välillä
print(array1 * array2)
```

```python
#jos muutetaan array1 arvoja siten että sen shape on  (2,4) eli 2 row, 4 column, tämä ei enää toimi!
array1 = np.array([[1,2,3,4],[5,6,7,8]])
print(array1.shape) # (2, 4) ei toimi, taulukoiden row/column arvot täytyy joko olla samat tai
print(array2.shape) # (4, 1) arvon täytyy olla 1 toisessa sarakkeista
```

```python
print(array1 * array2) # saadaan virhe ValueError: operands could not be broadcast together with shapes (2,4) (4,1) 
```

```python
#we can also reshape our arrays using reshape(argument1, argument2)
array_2d = np.array([[1,2,3],[4,5,6]]) #array shaped(2,3)
reshaped = array_2d.reshape(3,2) # lets reshape it to form (3,2)
print(array_2d)
print(reshaped)

#array can also be flattened by using flatten()
flattened = array_2d.flatten() #shape (1,6)
print(flattened)
```
